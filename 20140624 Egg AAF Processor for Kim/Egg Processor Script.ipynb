```python
def get_next(iterable, current_element):
    
    if current_element in iterable:
        index = iterable.index(current_element)

        if index == len(iterable) - 1:
            next_index = 0
        else:
            next_index = index + 1

        return iterable[next_index]
    
    else:
        raise ValueError("element not in iterable")
```

```python
def get_next_box_position(current_box, current_row, current_col):
    """
    We assume that each box has 9 rows and 9 columns. The rows are letters A->I, the columns are numbers 1->9,
    and boxes are continuous integers starting at 1 through to infinity.
    """
    rows = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
    cols = [1, 2, 3, 4, 5, 6, 7, 8, 9]
    
    new_column = get_next(cols, current_col)
    
    if new_column == 1:
        new_row = get_next(rows, current_row)
    else:
        new_row = current_row
        
    if (current_row, current_col) == ('I', 9):
        new_box = current_box + 1
        
    else:
        new_box = current_box
        
    return (new_box, new_row, new_column)
```

```python
# Current box, row, col correspond to the position of the current last sample in the box.
current_box = 2
current_row = 'G'
current_col = 4

def generate_sample_box_pos_list(current_box, current_row, current_col, filename):
    new_samples = []
    infile = pd.read_csv(filename)
    for row in infile.iterrows():
        sample_id = row[1]['Sample ID']
        tubes_collected = row[1]['Tubes Collected']
        date_of_harvest = row[1]['Date of Harvest']
        for i in range(tubes_collected):
            new_sample_name = sample_id + " Tube " + str(i+1) + " of " + str(tubes_collected)
            new_box, new_row, new_col = get_next_box_position(current_box, current_row, current_col)
            new_samples.append((new_sample_name, date_of_harvest, new_box, new_row, new_col))

            current_box = new_box
            current_row = new_row
            current_col = new_col
            
    return new_samples
```

```python
new_samples = generate_sample_box_pos_list(current_box, current_row, current_col, 'AFF_test.csv')
```

```python
def generate_sample_box_pos_df(box_pos_list):
    new_df = pd.DataFrame(box_pos_list)
    new_df.columns = ['Sample ID', 'Date of Harvest', 'Box', 'Row', 'Column']
    
    return new_df
```
